```python
import pandas as pd
import numpy as np
import random as rnf
import matplotlib.pyplot as plt 
import matplotlib
from matplotlib import colors
```

```python
j1 = 0 # Cara
j2 = 1 # Sello
```

```python
# Muestras
ejm1 = 50
ejm2 = 100
ejm3 = 150
```

```python
#Funcion generar un numero Aleatorio
def gen_num():
    return rnf.random()
```

```python
#Funcion el cual se da la misma probabilidad a ambos jugadores
def get_label(x):
    if x>0 and x <= 0.50:
        return j1
    else:
        return j2
```

```python
# Con variable X llamamos a la funcion aletoria
x = gen_num()
```

```python
#se crea una matriz con elemnetos a cero y vaiables auxiliares
Volados = np.zeros((ejm2, 3))
auxj1 = 0
auxj2 = 0
ganaj1 = 0
ganaj2 = 0
c = 0
s = 0
c1 = 0
c2 = 0
```

```python
for i in range(ejm2):
    for j in range(3):
        Volados[i,j] = get_label(gen_num())
        if Volados[i,j] == j1:
            auxj1 = auxj1 + 1
            print("cara ",auxj1)
            c = c + 1
        elif Volados[i,j] == j2:
            auxj2 = auxj2 + 1
            print("sello ",auxj2)
            s = s + 1
    if auxj1 > auxj2:
        print("-----Ronda %s gana Carlos Jugador 1"%(i))
        ganaj1 = ganaj1 + 1
    elif auxj2 > auxj1:
        print("-----Ronda %s gana Alexander Jugador 2" %(i))
        ganaj2 = ganaj2 + 1
    auxj1 = 0
    auxj2 = 0
```

```python
print("Alexander")
print("   Ha ganado ",ganaj1," rondas, la probalidad es: %s" %(ganaj1/ejm1))
print("Carlos")
print("   Ha ganado ",ganaj2," rondas, la probalidad es: %s" %(ganaj2/ejm1))

print("Caras salieron: ",c)
print("Sellos salieron: ",s)
```

```python
#Graficacion de barras 
fig, ax = plt.subplots()

carag = [ [c], [s] ]
    
Xs = np.arange(1) 
plt.bar(Xs + 0.00, carag[0], color = "r", width = 0.50)
plt.bar(Xs + 0.25, carag[1], color = "b", width = 0.50)

plt.xticks(Xs+0.10, [ejm2])
ax.set_title('Alexander VS Carlos')
ax.set_ylabel('Porcentaje')
ax.set_xlabel('Rondas')
plt.plot(x, x,'r', label='Carlos')
plt.plot(x, x**2,'b', label='Alexander')
plt.legend()
plt.show()
```
